```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import CSV
creditos_string = '../Resources/DatosCreditos.csv'
datos_creditos_df = pd.read_csv(creditos_string)
datos_creditos_df
```

```python
# Delete first row
tipos_creditos = datos_creditos_df[["Type of Credit","Total"]]

# Set the index to "Type of Credit" to be used as labels
tipos_creditos = tipos_creditos.set_index("Type of Credit")

tipos_creditos.head()
```

```python
# Create a bar chart
tipos_creditos.plot(kind="bar", figsize=(9,9), rot=90, color="orange")

label=tipos_creditos["Total"].tolist()

f = .25
for i in range(8):
    plt.text(x = f-0.5 , y = int(label[i])+0.1, s = label[i], size = 8, color="black")
    f = f + 1
    
plt.title("Type of Credits acquired by Mexico City's Population")
plt.show()
plt.tight_layout()
```

```python
# Create a pie chart 
plot = tipos_creditos.plot.pie(y='Total', figsize=(8, 11),autopct='%1.1f%%')
plt.legend(loc="best")
plt.title("Type of Credits acquired by Mexico City's Population")
```
