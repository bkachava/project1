```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Statistical Directory of Economic Units
# Directorio Estadístico Nacional de Unidades Económicas (DENUE) con datos de identificación, ubicación, actividad económica y tamaño de las unidades económicas activas en la CDMX
# Update 11/2018 Available in: http://www.beta.inegi.org.mx/contenidos/masiva/denue/denue_09_csv.zip
# In the repository the csv file is stored as a zip file to avoid conflicts with Git

# Save path to data set in a variable
data_file = "../Resources/denue_inegi_09_.csv"

# Use Pandas to read data
df_unieco_pd = pd.read_csv(data_file, low_memory=False) 

# Show the results
df_unieco_pd.head()
```

```python
# North America Industrial Classification System - Sistema de Clasificación Industrial de América del Norte 2018 (SCIAN 2018)
# Available in:  https://www.inegi.org.mx/contenidos/app/scian/SCIAN2018_Categorias-productos.xlsx
# The csv file is in the folder Resources

# Save path to data set in a variable
data_file = "../Resources/SCIAN2018_sector_2.csv"

# Use Pandas to read data
df_scian = pd.read_csv(data_file)

# Convert the two digits classification (sector) to six digits (class)
df_scian["codigo_act"] = (df_scian["codigo_act"] * 10000) + 9999

# Show the results
df_scian
```

```python
# Show the details of the dataframe
df_unieco_pd.info()
```

```python
# Show the different values of the field per_ocu
df_unieco_pd["per_ocu"].value_counts()
```

```python
# Create a dataframe with certain columns and only the data for three categories 
df_var_mas50 = df_unieco_pd.loc[(df_unieco_pd["per_ocu"] == "51 a 100 personas") | \
                                (df_unieco_pd["per_ocu"] == "101 a 250 personas") | \
                                (df_unieco_pd["per_ocu"] == "251 y más personas"), \
                                ["codigo_act", "nombre_act", "per_ocu", "tipo_vial", \
                                 "cod_postal", "cve_mun", "municipio", "latitud", "longitud"]]
# Show the results
df_var_mas50.head()
```

```python
# Show the counters for the data
df_var_mas50.count()
```

```python
# Update the value of the column per_ocu_min or per_ocu_max to calculate the total of occupied persons
df_var_mas50.loc[df_var_mas50["per_ocu"] == "251 y más personas", 'per_ocu_min'] = 251
df_var_mas50.loc[df_var_mas50["per_ocu"] == "251 y más personas", 'per_ocu_max'] = 252
```

```python
# Update the value of the column per_ocu_min or per_ocu_max to calculate the total of occupied persons
df_var_mas50.loc[df_var_mas50["per_ocu"] == "101 a 250 personas", 'per_ocu_min'] = 101
df_var_mas50.loc[df_var_mas50["per_ocu"] == "101 a 250 personas", 'per_ocu_max'] = 250
```

```python
# Update the value of the column per_ocu_min or per_ocu_max to calculate the total of occupied persons
df_var_mas50.loc[df_var_mas50["per_ocu"] == "51 a 100 personas", 'per_ocu_min'] = 51
df_var_mas50.loc[df_var_mas50["per_ocu"] == "51 a 100 personas", 'per_ocu_max'] = 100
```

```python
# Show the results
df_var_mas50.head()
```

```python
# Show how many unique values has the field
len(df_var_mas50["nombre_act"].unique())
```

```python
# Show the unique values of the field
df_var_mas50["nombre_act"].unique()
```

```python
# Show the different values of the field
df_var_mas50["nombre_act"].value_counts()
```

```python
# Create bins in which to place values for the sector
bins = df_scian["codigo_act"].tolist()
bins.insert(0, 0)
bins
```

```python
# Create labels for these bins
group_labels = df_scian["tit_act"].tolist()
group_labels
```

```python
# Place the data series into a new column inside of the DataFrame
df_var_mas50["Sector"] = pd.cut(df_var_mas50["codigo_act"], bins, labels=group_labels)
df_var_mas50.tail()
```

```python
# Show the different values of the field Sector
df_var_mas50["Sector"].value_counts()
```

```python
# Create a dataframe with the number of "unidades económicas" by municipio 
df_mun = pd.DataFrame(df_var_mas50["municipio"].value_counts())

# Reset index and rename the columns
df_mun.reset_index(inplace=True)
df_mun = df_mun.rename(columns={"index": "municipio", "municipio": "count"})

# Show results
df_mun
```

```python
# Create a group by municipality
mun_group = df_var_mas50.groupby("municipio")
```

```python
# Create a dataframe to calculate the minimum of occupied persons
df_min = mun_group["per_ocu_min"].sum()
smin = pd.DataFrame(df_min)
smin.reset_index(inplace=True)

# Show the columns
smin.columns
```

```python
# Create a dataframe to calculate an approximate of maximum of occupied persons
df_max = mun_group["per_ocu_max"].sum()
smax = pd.DataFrame(df_max)
smax.reset_index(inplace=True)

# Show the columns
smax.columns
```

```python
# Merge the dataframe of economic units with the minimum
df = df_mun.merge(smin, how='inner', on='municipio')

# Show the results
df
```

```python
# Merge the dataframe of economic units with the minimum and the maximum
df_mun_per = df.merge(smax, how='inner', on='municipio')

# Show the results
df_mun_per
```

```python
# Use a specific syle for the plot
with plt.style.context('fivethirtyeight', after_reset=True):
    
    # Set values for the plot properties
    plt.figure(figsize=(20,12))
    plt.xticks(rotation=45)   #'vertical'
    plt.xlim(-.5, 16)
    plt.axhline(y = 0, color = 'black', linewidth = 5, alpha = .7)
    plt.text(x = -1, y = -350,
        s = '__________________________________________________________________________________________________________________________________________________________________________________________________',
        color = 'grey', alpha = .7)
    plt.text(x = -.5, y = -410,
        s = 'Pytheam                                                                                                                                 Source: Own calculations based on data from INEGI - Statistical Directory of Economic Units',fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey')
    plt.title("Economic Units - Size of establishment > 51 persons", color = 'black', fontsize=20)
    plt.ylabel("Total", fontsize=16)

    # Create labels
    label = df_mun_per["count"].tolist()
    
    # Plot the values
    plt.bar(df_mun_per["municipio"], df_mun_per["count"], color="darkorange", align="edge")   #darkorange lightsteelblue

    # Text on the top of each barplot
    f = .7
    for i in range(16):
        plt.text(x = f-0.5 , y = int(label[i])+0.1, s = label[i], size = 14, color="black")
        f = f + 1

    # Save the Figure
    plt.tight_layout()
    plt.savefig("../Output/EconomicUnits.png")

    # Show the Figure
    plt.show()
```

```python
# Create a dataframe with the number of "unidades económicas" by municipio and sector
df_secmun = pd.DataFrame(df_var_mas50.groupby(['municipio', 'Sector'])["codigo_act"].count())

# Reset index
df_secmun.reset_index(inplace=True)

# Show results
df_secmun.columns
df_secmun
```

```python
# Save the dataframe in a csv for later use
df_var_mas50.to_csv("../Output/mas50.csv", encoding="ISO-8859-1", index=False, header=True)
```

```python
# Dependencies
import gmaps

# Google developer API key
gkey = "YOUR APY KEY GOES HERE"

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Select only the economic unit of one municipality
xochimilco_df = df_var_mas50[df_var_mas50["municipio"] == "Xochimilco"]

# Show the number of points to graph
xochimilco_df.count()
```

```python
# Store latitude and longitude in locations
locations = xochimilco_df[["latitud", "longitud"]].head()

# Show the locations
locations
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer 
heat_layer = gmaps.heatmap_layer(locations, dissipating=False, max_intensity=10, point_radius=1)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# Customize the size of the figure
figure_layout = {
    'width': '400px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Assign the marker layer to a variable
markers = gmaps.marker_layer(locations)

# Add the layer to the map
fig.add_layer(markers)

# Display figure
fig
```
